```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#read csv files into pandas dataframe
movies_df = pd.read_csv('movies.csv')
ratings_df = pd.read_csv('ratings.csv')
movies_df = movies_df[['movieId','title']]
```

```python
#no duplicated values. good.
ratings_df[ratings_df.duplicated(['userId','movieId'],keep=False)]
```

```python
#unify the movieIds
ratings_df.loc[ratings_df['movieId']==144606,['movieId']] = 6003
ratings_df.loc[ratings_df['movieId']==26958,['movieId']] = 838
ratings_df.loc[ratings_df['movieId']==147002,['movieId']] = 32600
ratings_df.loc[ratings_df['movieId']==168358,['movieId']] = 2851
ratings_df.loc[ratings_df['movieId']==64997,['movieId']] = 34048
```

```python
#join two tables to make the fact table
movies_ratings_df = pd.merge(ratings_df, movies_df, on ='movieId')
movies_ratings_df = movies_ratings_df[['userId','rating','title']]
```

```python
rating_counts = movies_ratings_df.groupby(['title'])['userId'].count().reset_index(name='rating_count')
```

```python
movies_ratings_df = pd.merge(movies_ratings_df,rating_counts,on='title')
```

```python
movies_averages = movies_ratings_df.groupby(['title'])['rating'].mean().reset_index(name='rating_average')
```

```python
movies_ratings_df = pd.merge(movies_ratings_df,movies_averages,on='title')
```

```python
movies_ratings_df
```

```python
sns.jointplot(x='rating_average',y='rating_count', data=movies_ratings_df.drop_duplicates('title'), alpha=0.5)
```

```python
movies_ratings_matrix = movies_ratings_df.pivot_table(values='rating' , index='userId' , columns='title' )
```

```python
toystory = movies_ratings_matrix['Toy Story (1995)']
```

```python
similar_to_toystory = movies_ratings_matrix.corrwith(toystory).reset_index(name='cor')
similar_to_toystory.dropna(inplace=True)
similar_to_toystory = pd.merge(rating_counts,similar_to_toystory,on='title')
```

```python
similar_to_toystory[similar_to_toystory['rating_count']>=50].sort_values('cor',ascending=False)
```

```python
def movies_recommender(movie_name, min_movie_rating):
    
    #get all the ratings of a movie which a user just watched
    movie_rating = movies_ratings_matrix[movie_name]
    
    #calculate the correlation coefficients among the ratings
    similar_movies =  movies_ratings_matrix.corrwith(movie_rating).reset_index(name='cor')
    similar_movies.dropna(inplace=True)
    
    #get the number of ratings voted
    similar_movies = pd.merge(rating_counts,similar_movies,on='title')
    
    return similar_movies[similar_movies['rating_count']>=min_movie_rating].sort_values('cor',ascending=False).head(20)
```

```python
movies_recommender('Toy Story (1995)',50)
```

```python
movies_recommender('Arachnophobia (1990)',50)
```

```python
movies_recommender('Juno (2007)',50)
```
